```python
import pandas as pd

# Ganti 'nama_file.csv' dengan path file CSV Anda
FileA = 'data_produk.csv'
FileB = 'jogjarobotika_scrap.csv'
# Memuat data
data = pd.read_csv(FileA)
data2 = pd.read_csv(FileB, sep=';')

data.head()


```

```python
data2.head()
```

```python
# dan menambahkan kolom 'kategori' dari 'data2' ke 'data'
data_merged = pd.merge(data, data2[['Judul', 'Katagory']], left_on='nama', right_on='Judul', how='left')

# Menghilangkan kolom 'judul' yang duplikat jika tidak diperlukan
data_merged.drop('Judul', axis=1, inplace=True)
```

```python
data_merged.head(50)
```

```python
kategori_unik = data_merged['Katagory'].unique()

# Menampilkan nilai unik
print("Kategori Unik:")
print(kategori_unik)
```

```python
data_grouped = data_merged.groupby('Katagory')
```

```python
data_grouped.head(50)
```

```python
# Mengurutkan data_merged berdasarkan 'kategori'
data_sorted = data_merged.sort_values(by='Katagory')

# Menyimpan hasil ke file CSV
data_sorted.to_csv('data_sorted_berdasarkan_kategori.csv', index=False)
```

```python
data_sorted['terjual'].head(25)
```

```python
# Fungsi untuk mengubah data 'terjual' ke numerik
def convert_to_number(s):
    if 'tidak tersedia' in s:
        return 0  # atau pd.NA untuk nilai NaN
    s = s.replace('rb+', '000').replace('+', '')  # Mengubah 'rb' menjadi '000' dan menghapus '+'
    s = ''.join(filter(str.isdigit, s))  # Menghapus semua karakter non-angka
    return int(s) if s else 0

# Menerapkan fungsi konversi ke kolom 'terjual'
data_sorted['terjual'] = data_sorted['terjual'].apply(convert_to_number)

# Menampilkan hasil
print(data_sorted['terjual'])

# Sekarang Anda bisa melakukan sorting atau perhitungan
# Misalnya, untuk mengurutkan data:
```
