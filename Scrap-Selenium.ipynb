```python
import csv
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

def scrape_toko(url_toko):
    # Inisialisasi WebDriver menggunakan GeckoDriver (Firefox)
    driver = webdriver.Firefox()

    try:
        # Buka URL
        driver.get(url_toko)

        # Tunggu hingga elemen-elemen produk muncul (gunakan XPath yang sesuai)
        produk_elements = WebDriverWait(driver, 80).until(
            EC.presence_of_all_elements_located((By.CLASS_NAME, 'css-1sn1xa2'))
        )

        # List untuk menyimpan data produk
        data_produk = []

        # Loop melalui setiap elemen produk dan mengekstrak informasi yang dibutuhkan
        for produk in produk_elements:
            try:
                nama_produk = produk.find_element(By.CLASS_NAME, 'prd_link-product-name').text.strip()
                harga_produk = produk.find_element(By.CLASS_NAME, 'prd_link-product-price').text.strip()
                link_produk = produk.find_element(By.CSS_SELECTOR, 'a.pcv3__info-content').get_attribute('href')
                foto_produk = produk.find_element(By.CSS_SELECTOR, 'img.css-1q90pod').get_attribute('src')

                # Mencari elemen berapa banyak produk terjual (jika ada)
                try:
                    jumlah_terjual_element = produk.find_element(By.CLASS_NAME, 'prd_label-integrity')
                    jumlah_terjual = jumlah_terjual_element.text.strip()
                except NoSuchElementException:
                    jumlah_terjual = "Tidak Tersedia"

                # Menambahkan data produk ke dalam list
                data_produk.append([nama_produk, harga_produk, link_produk, foto_produk, jumlah_terjual])

                # Menampilkan hasil
                print(f"Nama Produk: {nama_produk}")
                print(f"Harga: {harga_produk}")
                print(f"Link Produk: {link_produk}")
                print(f"URL Foto Produk: {foto_produk}")
                print(f"Jumlah Terjual: {jumlah_terjual}")
                print("=" * 50)

            except StaleElementReferenceException:
                print("StaleElementReferenceException pada elemen, melanjutkan ke elemen berikutnya.")
                continue

        # Simpan data ke dalam file CSV
        with open('produk_data4.csv', 'w', newline='', encoding='utf-8') as file:
            writer = csv.writer(file)
            # Menulis header
            writer.writerow(['Nama Produk', 'Harga', 'Link Produk', 'URL Foto Produk', 'Jumlah Terjual'])
            # Menulis data produk
            writer.writerows(data_produk)

    finally:
        # Tutup browser setelah selesai
        driver.quit()

# Ganti 'URL_TOKO' dengan URL toko yang ingin Anda scrape
URL_TOKO = 'https://www.tokopedia.com/jogjarobotika/product/page/4'
scrape_toko(URL_TOKO)
```

```python
import csv
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException

def wait_for_elements(driver, by, value, expected_count, timeout=10):
    try:
        WebDriverWait(driver, timeout).until(lambda d: len(d.find_elements(by, value)) >= expected_count)
        return driver.find_elements(by, value)
    except TimeoutException:
        return []

def scrape_data():
    # Inisialisasi driver Selenium untuk Firefox
    driver = webdriver.Firefox()

    # Masukkan URL dari halaman yang ingin di-scrape
    driver.get('https://www.tokopedia.com/jogjarobotika/product')

    data_produk = []
    produk_elements = wait_for_elements(driver, By.CLASS_NAME, 'css-1sn1xa2', len(data_produk) + 1, timeout=10)

    for produk in produk_elements[len(data_produk):]:
        try:
            nama_produk = produk.find_element(By.CLASS_NAME, 'prd_link-product-name').text.strip()
            harga_produk = produk.find_element(By.CLASS_NAME, 'prd_link-product-price').text.strip()
            link_produk = produk.find_element(By.CSS_SELECTOR, 'a.pcv3__info-content').get_attribute('href')
            foto_produk = produk.find_element(By.CSS_SELECTOR, 'img.css-1q90pod').get_attribute('src')

            # Mencari elemen berapa banyak produk terjual (jika ada)
            try:
                jumlah_terjual_element = produk.find_element(By.CLASS_NAME, 'prd_label-integrity')
                jumlah_terjual = jumlah_terjual_element.text.strip()
            except NoSuchElementException:
                jumlah_terjual = "Tidak Tersedia"

            data_produk.append({
                'nama': nama_produk,
                'harga': harga_produk,
                'link': link_produk,
                'foto': foto_produk,
                'terjual': jumlah_terjual
            })

        except Exception as e:
            print(f"Error: {e}")

    driver.quit()
    return data_produk

def save_to_csv(data_produk, filename):
    keys = data_produk[0].keys()
    with open(filename, 'w', newline='', encoding='utf-8') as output_file:
        dict_writer = csv.DictWriter(output_file, keys)
        dict_writer.writeheader()
        dict_writer.writerows(data_produk)

# Panggil fungsi scrape_data
produk_data = scrape_data()

# Simpan data ke CSV
save_to_csv(produk_data, 'data_produk.csv')
```

```python
import csv
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException

def scroll_down(driver):
    """Scroll down the page."""
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)  # Tunggu halaman memuat

def wait_for_elements(driver, by, value, expected_count, timeout=10):
    try:
        WebDriverWait(driver, timeout).until(lambda d: len(d.find_elements(by, value)) >= expected_count)
        return driver.find_elements(by, value)
    except TimeoutException:
        return []

def scrape_data(base_url, total_pages):
    # Inisialisasi driver Selenium untuk Firefox
    driver = webdriver.Firefox()

    data_produk = []

    for page in range(1, total_pages + 1):
        # Membuka halaman URL
        driver.get(f"{base_url}/page/{page}")
        time.sleep(2)  # Tunggu halaman untuk memuat

        scroll_down(driver)
        produk_elements = wait_for_elements(driver, By.CLASS_NAME, 'css-1sn1xa2', 1, timeout=10)

        for produk in produk_elements:
            try:
                nama_produk = produk.find_element(By.CLASS_NAME, 'prd_link-product-name').text.strip()
                harga_produk = produk.find_element(By.CLASS_NAME, 'prd_link-product-price').text.strip()
                link_produk = produk.find_element(By.CSS_SELECTOR, 'a.pcv3__info-content').get_attribute('href')
                foto_produk = produk.find_element(By.CSS_SELECTOR, 'img.css-1q90pod').get_attribute('src')

                try:
                    jumlah_terjual_element = produk.find_element(By.CLASS_NAME, 'prd_label-integrity')
                    jumlah_terjual = jumlah_terjual_element.text.strip()
                except NoSuchElementException:
                    jumlah_terjual = "Tidak Tersedia"

                data_produk.append({
                    'nama': nama_produk,
                    'harga': harga_produk,
                    'link': link_produk,
                    'foto': foto_produk,
                    'terjual': jumlah_terjual
                })

            except Exception as e:
                print(f"Error: {e}")

    driver.quit()
    return data_produk

def save_to_csv(data_produk, filename):
    keys = data_produk[0].keys()
    with open(filename, 'w', newline='', encoding='utf-8') as output_file:
        dict_writer = csv.DictWriter(output_file, keys)
        dict_writer.writeheader()
        dict_writer.writerows(data_produk)

# URL dasar (tanpa nomor halaman)
base_url = 'https://www.tokopedia.com/jogjarobotika/product'

# Total jumlah halaman
total_pages = 54

# Panggil fungsi scrape_data
produk_data = scrape_data(base_url, total_pages)

# Simpan data ke CSV
save_to_csv(produk_data, 'data_produk.csv')
```
